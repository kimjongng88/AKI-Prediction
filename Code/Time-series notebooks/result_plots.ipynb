```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the plot
time_windows = ['1 hour', '3 hours', '6 hours', '12 hours']
auc_59_features = [0.624, 0.808, 0.509, 0.513]
f1_score_59_features = [0.397, 0.762, 0.037, 0.049]
auc_39_features = [0.614, 0.672 ,0.532 ,0.504]
f1_score_39_features = [0.369 ,0.510 ,0.119 , 0.017]

# Setting the width and spacing of the bars
width = 0.35
spacing = 0.1

# Creating the x-axis positions for the bars
x = np.arange(len(time_windows)) # the label locations
x1 = x - width/2 - spacing/2 # for AUC - 59 features
x2 = x + width/2 + spacing/2 # for AUC - 39 features

# Creating the figure and the axes for AUC plot
fig_auc, ax_auc = plt.subplots()
ax_auc.bar(x1, auc_59_features, width, label='59 features')
ax_auc.bar(x2, auc_39_features, width, label='39 features')
ax_auc.set_xlabel('Time Window')
ax_auc.set_ylabel('AUC Score')
ax_auc.set_title('AUC Score by Number of Features')
ax_auc.set_xticks(x)
ax_auc.set_xticklabels(time_windows)
ax_auc.legend()

# Creating the x-axis positions for the bars
x3 = x - width/2 - spacing/2 # for F1 Score - 59 features
x4 = x + width/2 + spacing/2 # for F1 Score - 39 features

# Creating the figure and the axes for F1 Score plot
fig_f1, ax_f1 = plt.subplots()
ax_f1.bar(x3, f1_score_59_features, width, label='59 features')
ax_f1.bar(x4, f1_score_39_features, width, label='39 features')
ax_f1.set_xlabel('Time Window')
ax_f1.set_ylabel('F1 Score')
ax_f1.set_title('F1 Score by Number of Features')
ax_f1.set_xticks(x)
ax_f1.set_xticklabels(time_windows)
ax_f1.legend()

# Adjusting the layout and showing the plots
fig_auc.tight_layout()
fig_f1.tight_layout()
plt.show()
```
